## 16

```python
# 分割したい数N
N = int(input())
# データフレームの行数
n = len(df)

# N件ごとに取得する
splited_df = [df.loc[i:i+N-1, :] for i in range(0, n, N)]
for df_i in splited_df:
    display(df_i)  # NOTE: display()はjupyter内でのみ有効
```
